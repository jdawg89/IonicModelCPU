```matlab
ito1 = 2.6;
ito2 = ito1;
dx1 = 7;
dx2 = dx1;
str = ['LR2allap_PCL_1000_dx1_',num2str(dx1),'_dx2_',num2str(dx2),'_ito1_',num2str(ito1),'_ito2_',num2str(ito2),'.txt'];
str2 = ['LR2allca_PCL_1000_dx1_',num2str(dx1),'_dx2_',num2str(dx2),'_ito1_',num2str(ito1),'_ito2_',num2str(ito2),'.txt'];
%str = ['LR2allap_PCL_1000_dx1_',num2str(dx1),'_dx2_',num2str(dx2),'_ito1_',num2str(ito1),'_ito2_',num2str(ito2),'.txt'];
A = dlmread(str);
B = dlmread(str2);
counter1 = 0;
counter2 = 0;
counter3 = 0;
counter4 = 0;
cellnum1 = 1;
cellnum2 = 128; %100; %128;
cellnum3 = 16384 - 127; %200; %16384 - 127;
cellnum4 = 16384; %300; %16384;
times1 = [];
times2 = [];
times3 = [];
times4 = [];
len = length(A(2:end,cellnum1+1));
for i = 1:(len-1)
    if (A(i+1,cellnum1+1) < -75.0 && A(i+2,cellnum1+1) > -75.0)
        counter1 = counter1 + 1;
        times1 = [times1; A(i+1,1)];
    end
    if (A(i+1,cellnum2+1) < -75.0 && A(i+2,cellnum2+1) > -75.0)
        counter2 = counter2 + 1;
        times2 = [times2; A(i+1,1)];
    end
    if (A(i+1,cellnum3+1) < -75.0 && A(i+2,cellnum3+1) > -75.0)
        counter3 = counter3 + 1;
        times3 = [times3; A(i+1,1)];
    end
    if (A(i+1,cellnum4+1) < -75.0 && A(i+2,cellnum4+1) > -75.0)
        counter4 = counter4 + 1;
        times4 = [times4; A(i+1,1)];
    end
end
disp([ito1, counter1, counter2, counter3, counter4]);
```
