```matlab
apd = dlmread('LR2bifsPCL_500_varyisk_skh.txt');
isks = unique(apd(:,2));
skhs = unique(apd(:,3));
A = zeros(length(isks),length(skhs));
for i = 1:length(isks)
    isk = isks(i);
    for j = 1:length(skhs)
        skh = skhs(j);
        apdcur = abs(diff(apd(apd(:,2)==isk & apd(:,3)==skh,4:end)));
        if (max(apdcur) > 5) 
            A(i,j) = 1;
        else
            A(i,j) = 0;
        end
    end
end
contourf(isks,skhs,A')
A = [0.005*isks, A];
A = [[0, 1000*skhs']; A];
dlmwrite('LR2stab_PCL_500_varyisk_skh.txt',A,'delimiter','\t');
        
```
